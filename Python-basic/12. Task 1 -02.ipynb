For example:
Input:

Output:

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here
# Input for first fraction
num1 = int(input("Enter numerator of 1st fraction: "))
den1 = int(input("Enter denominator of 1st fraction: "))

# Input for second fraction
num2 = int(input("Enter numerator of 2nd fraction: "))
den2 = int(input("Enter denominator of 2nd fraction: "))

# Calculate the numerator and denominator of the sum
numerator = num1 * den2 + num2 * den1
denominator = den1 * den2

print("Sum of the fractions = ", numerator,"/",denominator)
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
# Milk tank dimensions input
H = int(input("Enter the height of the tank (in cm): "))
L = int(input("Enter the length of the tank (in cm): "))
B = int(input("Enter the breadth of the tank (in cm): "))

# Glass dimensions input
h = int(input("Enter the height of the glass (in cm): "))
r = int(input("Enter the radius of the glass (in cm): "))


# Volumes
volume_tank = H * L * B
volume_glass = 3.14 * r * r * h

# Number of glasses
num_glasses = volume_tank // volume_glass

print("Number of glasses of milk:",num_glasses)
````
